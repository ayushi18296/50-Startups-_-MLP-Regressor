```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

import io
A = pd.read_csv(io.BytesIO(uploaded['50_Startups.csv']))

A.head()
```

```python
 Y = A[["PROFIT"]]
```

```python
X = A[["MKT","RND","ADMIN"]]
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler(feature_range=(0,1))
X = pd.DataFrame(mm.fit_transform(X),columns=["MKT","RND","ADMIN"])
X = X.join(pd.get_dummies(A.STATE))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=33)
```

```python
from keras.models import Sequential
from keras.layers import Dense
nn = Sequential()
nn.add(Dense(32,input_dim = 6,activation = "relu"))
nn.add(Dense(32,activation="relu"))
nn.add(Dense(1))
```

```python
nn.compile(metrics="mse",optimizer="SGD",loss="MAE")
```

```python
nn.fit(xtrain,ytrain, validation_split=0.2,batch_size=30,epochs=2000)
```

```python
pred = nn.predict(xtest)
ytest
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(ytest,pred)
```
